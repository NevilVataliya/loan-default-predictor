```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('E:/Projects/Learn ML/loan-default-predictor/data/3_removed_outliers.csv')
```

```python
df.columns
```

```python
categorical_cols = [x for x in df.columns if df[x].dtype=='O']
```

```python
categorical_cols
```

```python
categorical_cols.remove('loan_status')
categorical_cols.remove('issue_d_date')
```

```python
for x in categorical_cols:
    print(f'{x}: {df[x].nunique()}')
```

```python
for x in categorical_cols:
    print(f'{x}:\t {df[x].value_counts()}')
```

### we can go with Label(one hot) Encoding for all categorical columns as they have less number of unique values

```python
df
```

```python
encoded_df = pd.get_dummies(df, columns=categorical_cols)
```

```python
encoded_df.to_csv("E:/Projects/Learn ML/loan-default-predictor/data/4_encoded_data.csv", index=False)
```
