## Removing Missing Values

```python
(useful_df.isnull().mean()*100).sort_values(ascending=False)
```

```python
pd.options.mode.copy_on_write = True
```

```python
## assuming for emp_lenth, who doesn't fill it are, unemployed, so fill with 0
useful_df['emp_length'] = useful_df['emp_length'].fillna(0)
```

```python
## dti abd revol_util are critical feature, guessing may cause bad impact, and it has very few rows so just drop it is the best idea
useful_df = useful_df.dropna(subset=['dti', 'revol_util'])
```

```python
## pub_rec_bankruptcies is rare feature, and almost all having value 0, and it's not having too much null values, so i can simply fill it with 0
useful_df['pub_rec_bankruptcies'] = useful_df['pub_rec_bankruptcies'].fillna(0)
```

```python
## for mort_acc, let me go with knn, to predict the specefic value for null mort_acc
# for this i have to check correlation with other cols

numeric_cols = [x for x in useful_df.columns if useful_df[x].dtype != 'O']
useful_df[numeric_cols].corr()['mort_acc'].sort_values(ascending=False)
```

```python
## let me take first 4 features and apply KNN to predict, most optimal mort_acc
mort_acc_correlated = useful_df[['total_acc', 'loan_amnt', 'annual_inc', 'revol_bal', 'mort_acc', 'open_acc']]
```

```python
mort_acc_correlated
```

```python
# print(mort_acc_correlated)
mort_train = mort_acc_correlated[mort_acc_correlated['mort_acc'].notna()]
mort_test = mort_acc_correlated[mort_acc_correlated['mort_acc'].isna()]
```

```python
from sklearn.neighbors import KNeighborsRegressor
classifier = KNeighborsRegressor()
```

```python
from sklearn.model_selection import RandomizedSearchCV
parameters = {
    "n_neighbors": [400,600,700]
}

knn_cv = RandomizedSearchCV(classifier, parameters, cv=5)
```

```python
mort_X_train = mort_train.drop(columns=['mort_acc'])
mort_y_train = mort_train['mort_acc']
mort_X_test = mort_test.drop(columns=['mort_acc'])
mort_y_test = mort_test['mort_acc']
```

```python
mort_y_train.value_counts()
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
mort_X_train = scaler.fit_transform(mort_X_train)
mort_X_test = scaler.transform(mort_X_test)
```

```python
knn_cv.fit(mort_X_train, mort_y_train)
```

```python
knn_cv.best_params_
```

```python
knn_cv.best_score_
```

```python
knn_train_pred = knn_cv.predict(mort_X_train)
from sklearn.metrics import r2_score
r2_score(mort_y_train, knn_train_pred)
```

#### 22% is to less, and i think our data having to much outlier, and also might be geometically data is not that much separated. so let's move to randomforest

```python
from sklearn.ensemble import RandomForestRegressor
rfc = RandomForestRegressor(n_jobs=4)
rfc.fit(mort_X_train, mort_y_train)
```

```python
mort_y_trpred = np.round(rfc.predict(mort_X_train)) # rounding because mort_acc is integer
```

```python
from sklearn.metrics import r2_score, mean_squared_error
print("R2:", r2_score(mort_y_train, mort_y_trpred))
print("MSE: ", mean_squared_error(mort_y_train, mort_y_trpred))
print("adjusted R2:", 1 - (1 - r2_score(mort_y_train, mort_y_trpred)) * (len(mort_y_train) - 1) / (len(mort_y_train) - mort_X_train.shape[1] - 1))
```

##### finally got 87%, because our data must have outliers, and KNN uses distance based approach so it is prone to outlier, while randomforest -> using decision tree will be robust for it 
##### now will predict missing mort_acc for nulls

```python
mort_y_testpred = np.round(rfc.predict(mort_X_test))
```

```python
mort_y_testpred
```

```python
useful_df.loc[mort_test.index, 'mort_acc'] = mort_y_testpred
```
