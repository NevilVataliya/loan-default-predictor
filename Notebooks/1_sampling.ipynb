```python
## sampling
mini_batches = []
chunk_size=50000

for chunk in pd.read_csv("/workspaces/codespaces-blank/data/Loan_status_2007-2020Q3.gzip", chunksize=chunk_size):
  chunk = chunk[chunk['loan_status'].isin(['Fully Paid', 'Charged Off', 'Default'])]
  if(not chunk.empty):
    sampled_chunk = chunk.groupby('loan_status').sample(frac=0.2, random_state=42)
    mini_batches.append(sampled_chunk)
```

```python
dataset = pd.concat(mini_batches)
```

```python
dataset
```

```python
dataset.drop(columns=['Unnamed: 0']).to_csv("/workspaces/codespaces-blank/data/sampled_data.csv", index=False)
```
